```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with Ramya 3C.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
data
```

```python
# pattern=r"\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}[\u202f[a-z]m|.*]\s-$"

# "24/03/2024, 9:38\u202fam"

pattern=r'\d{2}/\d{2}/\d{4}, \d{1,2}:\d{2}\s[ap]m -'
```

```python
messages=re.split(pattern,data)[1:]
len(messages)
```

```python
dates=re.findall(pattern,data)
```

```python
dates
```

```python
messages
```

```python
len(dates)
```

```python
data=pd.DataFrame({'user_message':messages,'message_date':dates})
data
```

```python
def change_timeframe(item):
    items=item.split()
    tm=items[-2]
    hr=int(items[1].split(":")[0])
    if(tm=='pm' and hr<12):
        t=items[1].split(':')
        num=int(t[0])+12
        t[0]=str(num)
        items[1]=":".join(t)
    return " ".join(items[:2])
data['time']=data['message_date'].apply(change_timeframe)
data['time']=pd.to_datetime(data['time'],format='%d/%m/%Y, %H:%M')
# data['message_date']=pd.to_datetime(data['message_data'],format="%d/%m/%Y, %H:%M 
```

```python
data
```

```python
# data.drop(columns=['message_date'],inplace=True)
data.rename(columns={'time':'date'},inplace=True)
data
```

```python
data.shape
```

```python
#separating users and messages

def get_user(item):
    s=item.split(":")
    if len(s)>1:
        return s[0];
    return ""
def get_message(item):
    s=item.split(":")
    if len(s)>=2:
        return ":".join(s[1:]);
    return s[0]
data['user']=data['user_message'].apply(get_user)
data['message']=data['user_message'].apply(get_message)
data
```

```python
data.drop(columns=['user_message'],inplace=True)
data
```
